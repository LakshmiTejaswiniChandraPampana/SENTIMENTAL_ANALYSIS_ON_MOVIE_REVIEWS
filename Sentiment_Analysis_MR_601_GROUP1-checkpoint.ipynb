# Text Preprocessing(for sentiment analysis)

Special characters are removed, and the text is converted to lowercase. Stopwords are eliminated, and stemming is applied to reduce words to their root form.

```python
# Download stopwords
import nltk
nltk.download('stopwords')
```

```python
# Preprocessing steps (removing special characters, stopwords, stemming)
ps=PorterStemmer()
corpus = []
for i in range(len(x)):
    review = re.sub( "[^a-zA-Z]"," ",x[i])
    review = review.lower()
    review = review.split()
    #removing stopwords from the data
    review = [ps.stem(word) for word in review if word not in set(stopwords.words("english"))]
    review = ' '.join(review)
    corpus.append(review)
```

# Model Inference

A function is created to predict the sentiment of new reviews, demonstrating the model's practical application.

```python
# Function to predict sentiment of a new review
def test_mod(sent):
    s=cv.transform([sent]).toarray()
    res=mnb.predict(s)[0]
    if res == 1:
        return "Positive"
    if res == 0:
        return "Negative"
```

# Additional Analysis

Word frequency analysis is conducted to identify common words in the reviews. Visualization tools like bar plots and word clouds are used to represent these findings effectively.

```python
# Combine all reviews in the corpus
all_words = ' '.join(corpus).split()

# Count the frequencies of each word
word_freq = Counter(all_words)

# Display the 20 most common words
common_words = word_freq.most_common(20)
print(common_words)

# Plotting the most common words
plt.figure(figsize=(10, 6))
sns.barplot(x=[word[0] for word in common_words], y=[word[1] for word in common_words])
plt.title('20 Most Common Words')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
# Function to plot top N-grams
def plot_top_ngrams(corpus, n=None, ngrams=2):
    vec = CountVectorizer(stop_words='english', ngram_range=(ngrams, ngrams)).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq = sorted(words_freq, key=lambda x: x[1], reverse=True)

    top_words = words_freq[:n]
    top_df = pd.DataFrame(top_words)
    top_df.columns=["Ngram", "Freq"]

    # Plot
    plt.figure(figsize=(10,6))
    sns.barplot(x="Freq", y="Ngram", data=top_df, palette="PuBuGn_d")
    plt.title(f'Top {ngrams}-grams')
    plt.show()

# Example usage: Plotting top 10 bi-grams
plot_top_ngrams(corpus, n=10, ngrams=2)
```
