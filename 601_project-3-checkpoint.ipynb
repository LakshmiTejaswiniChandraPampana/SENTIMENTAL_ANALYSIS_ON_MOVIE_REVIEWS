```python
import pandas as pd
import networkx as nx
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

# Assuming df is your DataFrame
df2 = df[[ 'Youtuber', 'rank', 'subscribers', 'video views','video_views_rank', 'category', 'uploads', 'Country', 'channel_type', 'channel_type_rank']]

# Create label encoders for 'country', 'category' and 'channel_type'
label_encoder_country = LabelEncoder()
label_encoder_category = LabelEncoder()
label_encoder_channel_type = LabelEncoder()

# Apply label encoding to the categorical columns
df2['country_encoded'] = label_encoder_country.fit_transform(df2['Country'])
df2['category_encoded'] = label_encoder_category.fit_transform(df2['category'])
df2['channel_type_encoded'] = label_encoder_channel_type.fit_transform(df2['channel_type'])

# Drop the original 'country', 'category' and 'channel_type' columns
df2.drop(columns=['Country', 'category', 'channel_type'], inplace=True)

# Create a graph for collaborations based on channel type or category
G = nx.Graph()

# Add nodes (YouTubers) to the graph
G.add_nodes_from(df2['Youtuber'])

# Iterate through the DataFrame to find potential collaborators
for i, row_i in df2.iterrows():
    for j, row_j in df2.iterrows():
        # Exclude the same YouTuber
        if i != j:
            # Customize the conditions based on your collaboration criteria
            # For example, you can collaborate if they share the same channel type or category
            if row_i['channel_type_encoded'] == row_j['channel_type_encoded'] or row_i['category_encoded'] == row_j['category_encoded']:
                # Add an undirected edge between YouTubers
                G.add_edge(row_i['Youtuber'], row_j['Youtuber'])

# Calculate network measures (you can add more as needed)
degree_centrality = nx.degree_centrality(G)
closeness_centrality = nx.closeness_centrality(G)

df2['degree_centrality'] = df2['Youtuber'].map(degree_centrality)
df2['closeness_centrality'] = df2['Youtuber'].map(closeness_centrality)


# Select only numeric columns for preprocessing
numeric_columns = df2.select_dtypes(include=['float64', 'int64']).columns
df2_numeric = df2[numeric_columns]

# Handle missing values if any
imputer = SimpleImputer(strategy='mean')
df2_imputed = pd.DataFrame(imputer.fit_transform(df2_numeric), columns=df2_numeric.columns)

# Standardize the data
scaler = StandardScaler()
df2_scaled = pd.DataFrame(scaler.fit_transform(df2_imputed), columns=df2_imputed.columns)

# Check the columns of df2_scaled
print(df2_scaled.columns)

# Create a user-item matrix (excluding the 'Youtuber' column if present)
if 'Youtuber' in df2_scaled.columns:
    user_item_matrix = df2_scaled.drop('Youtuber', axis=1).values
else:
    user_item_matrix = df2_scaled.values

# Calculate cosine similarity between Youtubers
similarity_matrix = cosine_similarity(user_item_matrix)
# Define alpha, beta, and gamma coefficients
alpha = 0.8  # Weight for content-based filtering
beta = 0.2   # Weight for collaborative filtering
gamma = 0.2  # Weight for network-based filtering
collab1=[]
def recommend_youtube1(youtube_name):
    yt_index = df2[df2['Youtuber'] == youtube_name].index[0]
    distances = similarity_matrix[yt_index]
    
    # Combine cosine similarity with network collaboration measures
    combined_scores = 0.4 * distances + 0.4 * df2['degree_centrality'] + 0.4 * df2['closeness_centrality']
    
    yt_list = sorted(list(enumerate(combined_scores)), reverse=True, key=lambda x: x[1])[1:9]
    for i, distance in yt_list:
        collab1.append(df2.iloc[i]['Youtuber'])
        print(df2.iloc[i]['Youtuber'])
        

recommend_youtube1('HYBE LABELS')
```
